```python
import cv2
```

```python
image_path = '/content/drive/MyDrive/SWM-1/DATASET/TEST/organic/O_12826.jpg'
img = cv2.imread(image_path)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#Displaying Image
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
#Loading Model

model = tf.keras.models.load_model('/content/drive/MyDrive/SWM-1/model_final_1.keras')
```

```python
model.summary()
```

## Testing Model

```python
image = tf.keras.preprocessing.image.load_img(image_path,target_size=(150,150))
input_arr = tf.keras.preprocessing.image.img_to_array(image)
input_arr = np.array([input_arr])
print(input_arr.shape)
```

```python
prediction = model.predict(input_arr)
prediction.shape
```

```python
result_index = np.argmax(prediction)
result_index
```

```python
class_name = ['organic', 'recyclable']
class_name[result_index]
```
